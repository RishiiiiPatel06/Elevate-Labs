4. remove rupee symbol in the rows of "discounted_price", "actual_price" and add rupee symbol in column name. also change data type to object to int.      

```python
# remove rupees symbol in the discounted_price column.
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

# remove rupees symbol in the actual_price column.
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹","",regex=False)
    .str.replace(",","",regex=False)
)

print(df.actual_price)
print(df.discounted_price)
```

```python
# rename the column of discoutned_price and acutal_price.

df = df.rename(columns={'discounted_price': 'discounted_price (₹)'})

df = df.rename(columns={"actual_price": "actual_price (₹)"})

```

```python
# change the datatype of the both columns.

df['discounted_price (₹)'] = df['discounted_price (₹)'].astype(float)

df["actual_price (₹)"] = df["actual_price (₹)"].astype(float)
```

##### 5. remove % symbol in the rows and change the datatype to object to int.

```python
# remove the % symbol in the column.
df["discount_percentage"] = (
    df["discount_percentage"].str.replace("%","",regex=False)
)

# change the column name.
df = df.rename(columns={"discount_percentage": "discount_percentage (%)"})

# change the data type of column.
df["discount_percentage (%)"] = df["discount_percentage (%)"].astype(float)
```

##### 6. change data type of rating and rating_count.

```python
# change datatype
df["rating_count"] = (
    df["rating_count"].astype(str).str.replace(",", "", regex=False)
)

df["rating_count"] = df["rating_count"].astype(float)
```

```python
df['rating'] = df['rating'].astype(str).str.replace('|', '', regex=False).str.strip()

df['rating'].info()
df['rating'].head()
```

##### 7. replace null values to NaN value and then remove it form the dataset.

```python
# Replace empty strings with NaN
df['rating'] = df['rating'].replace('', pd.NA)

# Drop rows with NaN in the 'rating' column
df = df.dropna(subset=['rating'])

# Convert the 'rating' column to float
df['rating'] = df['rating'].astype(float)
```
